# **Visualisasi Jaringan pada Jalur Transportasi Jawa Barat**

```
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from(["Bekasi", "Karawang", "Depok", "Bogor", "Sukabumi",
                  "Cianjur", "Cimahi", "Bandung", "Garut", "Purwakarta", "Subang",
                  "Sumedang","Indramayu","Majalengka", "Kuningan", "Ciamis", "Cirebon",
                  "Tasikmalaya"])
G.nodes()
nx.draw(G, with_labels = True)
fig = plt.figure(figsize=(20,50))
plt.show()
```

```
G.add_edge("Bekasi", "Karawang")
G.add_edge("Bekasi", "Depok")
G.add_edge("Karawang", "Purwakarta")
G.add_edge("Karawang", "Subang")
G.add_edge("Purwakarta", "Subang")
G.add_edge("Purwakarta", "Cimahi")
G.add_edge("Depok", "Bogor")
G.add_edge("Bogor", "Cianjur")
G.add_edge("Bogor", "Sukabumi")
G.add_edge("Cianjur", "Sukabumi")
G.add_edge("Cianjur", "Cimahi")
G.add_edge("Cimahi", "Bandung")
G.add_edge("Bandung", "Sumedang")
G.add_edge("Bandung", "Purwakarta")
G.add_edge("Sumedang", "Majalengka")
G.add_edge("Majalengka", "Indramayu")
G.add_edge("Indramayu", "Cirebon")
G.add_edge("Majalengka", "Cirebon")
G.add_edge("Cirebon", "Kuningan")
G.add_edge("Kuningan", "Ciamis")
G.add_edge("Majalengka", "Kuningan")
G.add_edge("Bandung", "Tasikmalaya")
G.add_edge("Bandung", "Ciamis")
G.add_edge("Tasikmalaya", "Ciamis")
G.add_edge("Tasikmalaya", "Garut")
G.add_edge("Garut", "Bandung")
G.add_edge("Kuningan", "Cirebon")
G.add_edge("Bandung", "Subang")
G.add_edge("Sukabumi", "Depok")
G.add_edge("Purwakarta", "Subang")
G.edges()
nx.draw(G, with_labels=True)
plt.show()
```

# **Penentuan Kota Terpenting dari Nilai Degree Centrality dan Betweenness Centrality**

## **Nilai Degree Centrality**

```
nx.degree_centrality(G)
```

Berdasarkan hasil dari nilai degree centrality, didapatkan nilai-nilai untuk setiap kota sehingga dapat diketahui 3 kota penting di Jawa Barat, yaitu :
1. Bandung,
dengan nilai DC = 0.4117647058823529
2. Majalengka, dengan nilai DC = 0.23529411764705882
3. Purwakarta, dengan nilai DC = 0.23529411764705882

## **Betweenness Centrality**

```
nx.betweenness_centrality(G)
```

Berdasarkan hasil dari nilai *betweenness centrality*, didapatkan nilai-nilai untuk setiap kota sehingga dapat diketahui 3 kota penting di Jawa Barat, yaitu :
1. Bandung,
dengan nilai BC = 0.5968137254901961
2. Cimahi, dengan nilai BC = 0.27818627450980393
3. Cianjur, dengan nilai BC = 0.21200980392156865
